### Импорт библиотек

```python
import matplotlib.pyplot as plt
import pandas as pd
```

### Функция чтения и обработки файла

```python
def text_processing(name:str):
    ident_str = ['Число эл-тов : ', 'Число сравнений : ', 
                'Число копирований ', 'Среднее время : ' ]
    with open(name) as f:
        text = f.readlines()
    size = []
    count_comp = []
    count_copy = []
    mean_time = []
    for i in text:
        if ident_str[0] in i:
            size.append(int(i[15:]))
        elif ident_str[1] in i:
            count_comp.append(int(i[17:]))
        elif ident_str[2] in i:
            count_copy.append(int(i[19:]))
        elif ident_str[3] in i:
            mean_time.append(float(i[15:]))
    return size, count_comp, count_copy, mean_time

```

### Делаем из списков обекты pd.Series

```python
def to_pd(name:str):
    size, count_comp, count_copy, mean_time = text_processing(name)
    comparisons_copy = pd.Series(dict(zip(size,count_copy )))
    comparisons_comp = pd.Series(dict(zip(size,count_comp )))
    comparisons_time = pd.Series(dict(zip(size,mean_time )))
    return comparisons_copy, comparisons_comp, comparisons_time
```

### Функция для построения графиков

```python
def do_graph(name:str, name_graph) -> None:    
    copy, comp, buble_time = to_pd(name)

    fig, axs = plt.subplots(nrows= 3 , ncols= 1 )

    axs[0].plot(copy)
    axs[0].set_title('Копирования')
    fig.subplots_adjust(top = 2)
    axs[1].plot(comp)
    axs[1].set_title('Сравнения')
    axs[2].set_title('Время')
    axs[2].plot(buble_time)
    print('Графики для сортировки', name_graph, ':')

```

# Строим графики
